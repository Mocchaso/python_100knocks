```python
# q97(q96)
# 完全数
"""
問題：高橋君は完全なのものが大好きです。
自然数には、完全数という物があります。完全数というのは、自分以外の約数の総和が自分と等しくなる自然数のことです。
例えば6の場合1+2+3=6となるので完全数です。それに対して、自分以外の約数の総和が自分より小さくなる場合は不足数と言い、大きくなる場合は過剰数と言います。
高橋君には今気になっている自然数があります。高橋君のために、それが完全数なのか不足数なのか過剰数なのか判定してください。
"""

## 自分の解法
print("自分の解法\n")

def divisor(n):
    # nの約数(自分自身、すなわちnを除く)を全て求める関数
    result = [i+1 for i in range(n) if n % (i+1) == 0]
    result.pop()
    return result

def check_perfect(n):
    if sum(divisor(n)) == n:
        return "完全数"
    elif sum(divisor(n)) < n:
        return "不足数"
    elif sum(divisor(n)) > n:
        return "過剰数"

print(check_perfect(6))

## 記事内の解法
print("\n記事内の解法\n")

def perfect_number(n):
    y_num = []
    for i in range(2, n):
        if n % i == 0:
            y_num.append(i)
            m = n // i
            y_num.append(m)
    y_num.append(1)
    y_num = set(y_num)
    sum_y = sum(y_num)
    if sum_y == n:
        print("完全数です")
    else:
        print("完全数ではありません")

perfect_number(6)
```

```python
# q98(q97)
# ハミング距離
"""
問題：2つの整数が与えられます。
これらを2進数に直した時の「ハミング距離」を測りなさい。
"""

## 自分の解法
print("自分の解法\n")

def hamming(x, y):
    xy_xor = x ^ y # xとyのビット排他的論理和
    return str(xy_xor).count("1")

print(hamming(10, 12)) # 2
print(hamming(17, 117))

## 記事内の解法(修正後)
print("\n記事内の解法(修正後)\n")

def hamming_distance(n, m):
    return format(n ^ m, 'b').count('1')

print(hamming_distance(10, 12)) # 2
print(hamming_distance(17, 117))
```

```python
# q100(q99)
# Number Base
"""
問題：ある進数(radix)で記された数字(str_number)が与えられます。
この数字を10進法に直して表示しなさい。
条件：そもそも与えられた数字が、与えられた進法で表すことができないものであった場合は"-1"と表示
例：
convert_10("AF", 16) == 175
convert_10("101", 2) == 5
convert_10("Z", 36) == 35
convert_10("AB", 10) == -1
"""

## 自分の解法
print("自分の解法\n")

import string

def convert_10(str_number, radix):
    nums = [str(i) for i in range(10)] # ["0", "1", "2", ... , "9"]
    nums.extend([i for i in string.ascii_uppercase]) # ["0", "1", ..., "9", "A", ... , "Z"]
    nums = dict(zip(nums, [i for i in range(36)])) # numsに通し番号を割り振った辞書を作ってnumsに上書き
    result = 0
    for i in range(len(str_number)):
        if nums[str_number[i]] + 1 > radix:
            # str_numberのいずれかがradix進法表記の範囲外ならreturn
            # +1することで、radixとnumsの各値の間にあるズレを直している
            # 例："9"までが10進法と判別できるが、+1しないと"A"までが10進法と判別してしまう
            return -1
        else:
            # str_numberの末尾から順に、進数変換の計算を行う
            result += (radix**i) * (nums[str_number[-(i+1)]])
    return result
    
print(convert_10("AF", 16)) # 175
print(convert_10("101", 2)) # 5
print(convert_10("Z", 36)) # 35
print(convert_10("AB", 10)) # -1
print(convert_10("3C5", 16)) # 965

## 記事内の解法
print("\n記事内の解法\n")

def convert_10_2(str_number, radix):
    try:
        return int(str_number, radix)
    except ValueError:
        return -1

convert_10_2("3C5",16)
```

```python
# q101(q100)
# Double Substring
"""
問題：あるテキストが与えられます。
そのテキストの内、ある文字列が2回登場した場合、その文字列の文字数を表示してください。
例：
double_substring('aaaa') ==> 2
double_substring('abc') ==> 0
double_substring('aghtfghkofgh') ==> 3 # fgh
"""

## 自分の解法
print("自分の解法\n")

def double_substring(text):
    count_dict = {}
    result = 0
    for ch in text:
        if not ch in count_dict:
            count_dict[ch] = 1
        else:
            count_dict[ch] += 1
            if count_dict[ch] == 2:
                result += 1
                count_dict[ch] = 0
    print(result)

double_substring('aaaa') # 2
double_substring('abc') # 0
double_substring('aghtfghkofgh') # 3(fgh)

## 記事内の解法
print("\n記事内の解法\n")

def double_substring2(line):
    s = []
    for i in range(len(line)-1) :
        for j in range(i+1,len(line)+1) :
            if line[i:j] in line[j:] :
                s.append(len(line[i:j]))

    if len(s)>0 :
        return max(s)
    else :
        return 0

print(double_substring2('aaaa')) # 2
print(double_substring2('abc')) # 0
print(double_substring2('aghtfghkofgh')) # 3(fgh)
```
