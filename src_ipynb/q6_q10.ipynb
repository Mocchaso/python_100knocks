```python
# q6
# FizzBuzz問題
"""
1～100までのうち、
3で割り切れるもの："Fizz!"
5で割り切れるもの："Buzz!"
15で割り切れるもの："FizzBuzz!"
それ以外：そのままの数
を表示させなさい
"""

for i in range(1, 101):
    if i % 15 == 0:
        print("FizzBuzz!")
    if i % 3 == 0:
        print("Fizz!")
    elif i % 5 == 0:
        print("Buzz!")
    else:
        print(i)

# リスト内包表記 -> if-elif-elseの場合は使用できない...
```

```python
# q7
# FizzBuzz問題の内、整数nまでの"z"の個数を計算し表示せよ
# (ただしif文,for文の使用不可)

# if文とfor文使ったやり方しか分からなかった...
def count_z(n):
    count = 0
    for i in range(1, n+1):
        if i % 15 == 0: # FizzBuzz
            count += 4
        elif i % 3 == 0: # Fizz
            count += 2
        elif i % 5 == 0: # Buzz
            count += 2
    return count

count_z(100)

# 記事内の解法

def count_z(n):
    print((n // 3 * 2) + n // 5 * 2) # 「//」は割り算の整数部分の結果を出す

count_z(100)
```

```python
# q8
# 為替自動換算クラスの作成
# 問題：日本円とドルとユーロに換算するクラスを作成せよ。
# 条件1：1ドル=109円、1ユーロ=129円で換算。(2018年5月8日時点)
# 条件2：クラスの引数に日本円を入力して、その後その値を各通貨に換算できるようにする。

# Pythonのクラスは、継承が無い場合、末尾に()は不要
# Pythonのコンストラクタは、def __init__(self)
class YenExchange():
    def __init__(self, yen):
        self.yen = yen
    
    def yen_to_dollar(self, yen):
        return self.yen / 109 
    
    def yen_to_euro(self, yen):
        return self.yen / 129

yen = 3000
exchange = YenExchange(yen)
print("{}円 = {}ドル".format(yen, exchange.yen_to_dollar(yen)))
print("{}円 = {}ユーロ".format(yen, exchange.yen_to_euro(yen)))

# 記事内の解法
print("記事内の解法")
class YenToCurrency:
    def __init__(self, yen):
        self.yen = yen
        
    def doll(self):
        doll = self.yen / 109
        return(doll)
    
    def euro(self):
        euro = self.yen / 129
        return(euro)

exchange = YenToCurrency(3000)
print("3000円は{}ドルです。".format(exchange.doll()))
print("3000円は{}ユーロです。".format(exchange.euro()))
```

```python
# q9
# RPGゲームクラスの作成
# 問題：キャラクターのステータスを登録して、お互いに攻撃することができるクラスを作成せよ。
# 条件1：キャラクターは名前、体力の現在値、体力の最大値、攻撃力、防御力の5つのパラメータを持っており、いつでも参照することができる。
# 条件2：キャラクターは別のキャラクターを攻撃して、相手の体力を自分の攻撃力(-相手のの防御力)分だけ減らすことができる。

class RPG_game:
    def __init__(self, name, hp, max_hp, atk, defe):
        self.name = name
        self.hp = hp
        self.max_hp = max_hp
        self.atk = atk
        self.defe = defe
    
    def attack(self, enemy):
        damage = self.atk - enemy.defe
        enemy.hp -= damage
        print("{}は{}に{}のダメージを与えた！".format(self.name, enemy.name, damage))
        print("{}の残りHP：{}".format(self.name, self.hp))
        print("{}の残りHP；{}".format(enemy.name, enemy.hp))
        
chara1 = RPG_game("もっちゃん", 8000, 10000, 500, 100)
chara2 = RPG_game("もっちゃそ", 10000, 10000, 700, 100)

chara2.attack(chara1)

# 記事内の解法
print("記事内の解法")
class Character:
    def __init__(self, name, maxhp, hp, attack_point, defense_point):
        self.name = name
        self.maxhp = maxhp
        self.hp = hp
        self.attack_point = attack_point
        self.defense_point = defense_point
    
    def status(self):
        return "{}：体力{}/{} 攻撃力{} 防御力{}".format(self.name, self.hp, self.maxhp, self.attack_point, self.defense_point)
    
    def attack(self, enemy):
        cal_attack_point = self.attack_point - enemy.defense_point
        enemy.hp -= cal_attack_point
        print("{}の攻撃！{}に{}のダメージ！".format(self.name, enemy.name, cal_attack_point))

yusha = Character("勇者", 60, 60, 10, 2)
slime = Character("スライム", 15, 15, 5, 1)

# ステータスを表示
print(yusha.status())
print(slime.status())

# 勇者の攻撃
yusha.attack(slime)
# スライムの攻撃
slime.attack(yusha)

# ステータスを表示
print(yusha.status())
print(slime.status())
```

```python
# q10
# Hashnumber判定
# 問題：整数Xの各桁の数字の和をf(X）としたとき、Xがf(X)で割り切れる場合、Xはハーシャッド数である。
# 整数Nが与えられるので、ハーシャッド数かどうかを判定せよ。

def judge_hash(n):
    str_n = str(n) # 整数Nを文字列に変換
    fx = 0
    for num in str_n: # Pythonの文字列は配列として扱える
        fx += int(num) # 各要素を整数に戻して順番に足していく(本当はsum使いたいけどうまく適用できなかった...)
    if n % fx == 0:
        print("{}はハーシャッド数である。".format(n))
    else:
        print("{}はハーシャッド数ではない。".format(n))

judge_hash(444)

# 記事内の解法
print("記事内の解法")

# list, mapを組み合わせてsumを活用しているところがミソ
def j_hash(n):
    s = str(n)
    array = list(map(int, list(s)))
    a = sum(array)
    if n % a == 0:
        print("Hashnumber")
    else:
        print("Not hashnumber")

j_hash(444)
```
