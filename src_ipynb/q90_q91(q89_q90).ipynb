```python
# q90(q89)
# 問題：渡辺君は、黒板に書かれている整数がすべて偶数であるとき、次の操作を行います。
# 黒板に書かれている整数すべてを、2で割ったものに置き換える。
# 渡辺君は最大で何回操作を行うことができるかを求めてください。
# 条件：数字はリスト型で与えられる。
# 例：count_odd([16, 12, 24]) ==> 2
# 1回操作を行うと[8, 6, 12]になります。2回操作を行うと[4, 3, 6]になります。2個目の3が奇数なため3回目の操作は行えません。

## 自分の解法
print("自分の解法\n")

count = 0
def count_divide(nums):
    global count
    print(nums)
    for i in range(len(nums)):
        if nums[i] % 2 == 1:
            return count
        nums[i] /= 2
    count += 1
    count_divide(nums)
    return count

print(count_divide([16, 12, 24]))

## 改善後
# map()でリスト内の数字を同時に2で割って、余りの合計が0でなかったら操作は行えないと判定する
print("\n改善後\n")

def count_devide2(init_list):
    count = 0
    tmp_list = init_list
    print(tmp_list)
    while sum(list(map(lambda x: x%2, tmp_list))) == 0: # tmp_list内の数字を2で割った余りの合計が0になったらループから抜ける
        tmp_list = list(map(lambda x: x/2, tmp_list)) # リスト内の数字を同時に2で割って更新する
        print(tmp_list)
        count += 1
    return count

init_list = [16, 12, 24]
print(count_devide2(init_list))

## 記事内の解法
print("\n記事内の解法\n")

def count_n(n_list):
    counts = []
    for n in n_list:
        count = 0
        while n % 2 == 0:
            count += 1
            n = n / 2
        else:
            counts.append(count)
    aws = min(counts)
    print(aws)

count_n([16,16,4])
```

```python
# q91(q90)
# 問題：
# N枚のカードがあり、0～Nまでの整数が被らないように書かれています。
# AliceとBobはこれらのカードを使ってゲームを行います。ゲームでは2人が交互に1枚ずつカードを取っていきます。Aliceが先にカードを取ります。
# 2人がすべてのカードを取ったときゲームは終了し、取ったカードの数の合計がその人の得点になります。
# 2人とも自分の得点を最大化するように最適戦略をとったとき、AliceはBobより何点多くの得点を獲得できるかを求めてください。

## 自分の解法
print("自分の解法\n")

def game(N):
    cards = [card for card in range(N+1)] # 0～NまでのN枚のカード
    print("カード：{}".format(cards))
    players = {"Alice": [], "Bob": []} # プレイヤーのキーに対応する値のリストに、そのプレイヤーが取ったカードを格納する
    turn = "Alice" # ターン変更、プレイヤーが取ったリストへの参照に使う
    while len(cards): # カードが無くなるまで
        taken_card = cards.pop(cards.index(max(cards)))
        players[turn].append(taken_card) # 現在のターンのプレイヤーがカードを取る
        turn = "Bob" if turn == "Alice" else "Alice" # ターン切り替え
    score_difference = sum(players["Alice"]) - sum(players["Bob"])
    print("Aliceの得点はBobより{}点高いです。".format(score_difference))
    return 0

game(10)

## 元の問題にチャレンジ
print("\n元の問題にチャレンジ\n")

import random

def game2(N):
    cards = [random.randrange(1, 100, 1) for card in range(N+1)] # 1～100の整数がランダムにN枚格納される
    print("カード：{}".format(cards))
    players = {"Alice": [], "Bob": []} # プレイヤーのキーに対応する値のリストに、そのプレイヤーが取ったカードを格納する
    turn = "Alice" # ターン変更、プレイヤーが取ったリストへの参照に使う
    while len(cards): # カードが無くなるまで
        taken_card = cards.pop(cards.index(max(cards)))
        players[turn].append(taken_card) # 現在のターンのプレイヤーがカードを取る
        turn = "Bob" if turn == "Alice" else "Alice" # ターン切り替え
    score_difference = sum(players["Alice"]) - sum(players["Bob"])
    print("Aliceの得点はBobより{}点高いです。".format(score_difference))
    return 0

game2(20)

## 記事内の解法
print("\n記事内の解法\n")

def num_fight(card_number):
    cards = list(range(card_number))
    count = 0
    alice_cards = []
    bob_cards = []
    while len(cards) != 0:
        taken_card = max(cards)
        cards.remove(taken_card)
        count += 1
        if count % 2 == 1:
            alice_cards.append(taken_card)
        else:
            bob_cards.append(taken_card)
    alice_score = sum(alice_cards)
    bob_score = sum(bob_cards)
    aws = alice_score - bob_score

    print(aws)

num_fight(10)
```
